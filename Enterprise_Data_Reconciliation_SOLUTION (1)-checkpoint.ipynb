# Enterprise Data Architecture & Reconciliation — COMPLETE SOLUTION

## The Scenario
You are a Senior Data Analyst at a massive FinTech company. The Data Engineering team is on vacation, and the CFO needs the Q1-Q4 Financial Audit Dashboard by Friday.

You have been given access to three separate, messy systems:
1. **The HR Database (`enterprise_database.db`):** A SQL database containing user accounts. *Warning: It contains historical logs, meaning users appear multiple times. You only want their most recent status.*
2. **The Server Logs (`server_logs.txt`):** 100,000 raw text logs. The transaction data is trapped inside strings.
3. **The Finance Server (`daily_exchange_rates.csv`):** A list of daily EUR to USD exchange rates. *Warning: The financial market is closed on weekends, so weekend dates have missing exchange rates.*

---

## Pipeline Architecture
```
enterprise_database.db  ──► [SQL Window Function] ──► df_active_users (14,183 rows)
                                                              │
server_logs.txt         ──► [Regex Extraction]    ──► extracted_data (94,902 rows)
                                                              │
daily_exchange_rates.csv ──► [ffill Weekend Gaps] ──► df_rates (365 rows, no gaps)
                                                              │
                              [Temporal Merge + Inner Join]   │
                                                              ▼
                                                     df_final (53,899 rows)
                                                              │
                                                     CFO Dashboard
```

````python
# ===================================================================
# IMPORTS — all libraries needed for the full pipeline
# ===================================================================
import sqlite3
import pandas as pd
import re
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import time
import warnings
warnings.filterwarnings('ignore')

pipeline_start = time.time()
print('Libraries loaded. Pipeline starting...')
````

---
## PHASE 1: Advanced SQL — Version-Controlled User Extraction

**The Trap:** The `users` table is a historical log — each user appears multiple times as their status changes. A naive `SELECT * FROM users WHERE status = 'Active'` would return users who are *currently* Suspended but were *previously* Active.

**The Solution:** A SQL **Window Function** using `ROW_NUMBER() OVER (PARTITION BY user_id ORDER BY updated_at DESC)` assigns rank 1 to the most recent record per user. We then filter for rank = 1 AND status = 'Active'.

````python
# ===================================================================
# PHASE 1: SQL EXTRACTION — Window Function to get latest user status
# ===================================================================

# Connect to the SQLite database
conn = sqlite3.connect('enterprise_database.db')

# Window Function query:
# - PARTITION BY user_id  → restart the rank counter for each unique user
# - ORDER BY updated_at DESC → most recent record gets rank 1
# - Outer WHERE filters to rank=1 (most recent) AND status='Active'
sql_query = """
    SELECT user_id, name, status, updated_at
    FROM (
        SELECT *,
               ROW_NUMBER() OVER (
                   PARTITION BY user_id
                   ORDER BY updated_at DESC
               ) AS rn
        FROM users
    )
    WHERE rn = 1
      AND status = 'Active'
"""

df_active_users = pd.read_sql_query(sql_query, conn)
conn.close()

print(f"✅ Phase 1 Complete: {len(df_active_users):,} currently-Active users extracted")
print(f"   Columns: {list(df_active_users.columns)}")
df_active_users.head()
````

---
## PHASE 2: Regex & Unstructured Data Wrangling

**The Trap:** Transaction data is buried inside raw log strings like:  
`[2023-03-15 00:00:00] INFO: Transaction processed. payload={"u":"U-19484", "item_code":"PRD-874", "eur_val":3404.44}`

**The Solution:**  
1. Drop `ERROR` rows (they have no `eur_val` field — corrupted).  
2. Use **vectorized** `str.extract()` with a single regex to pull all 4 fields in one pass — no `for` loops, O(n) performance on 100k rows.

````python
# ===================================================================
# PHASE 2: REGEX & UNSTRUCTURED DATA WRANGLING
# ===================================================================

# Load the raw server log file
df_logs = pd.read_csv('server_logs.txt')
print(f"   Raw log rows: {len(df_logs):,}")

# Step 1: Filter out ERROR rows (corrupted — no eur_val present)
df_logs = df_logs[~df_logs['Raw_Server_Log'].str.contains('ERROR', na=False)]
print(f"   After dropping ERROR rows: {len(df_logs):,}")

# Step 2: Vectorized regex extraction with str.extract() — NO for loops
# Regex breakdown:
#   \[(\d{4}-\d{2}-\d{2})   → capture the date from the timestamp bracket
#   "u":"(U-\d+)"           → capture the user ID
#   "item_code":"(PRD-\d+)" → capture the product ID
#   "eur_val":(\d+\.\d+)    → capture the Euro transaction value
pattern = r'\[(\d{4}-\d{2}-\d{2}).*?"u":"(U-\d+)".*?"item_code":"(PRD-\d+)".*?"eur_val":(\d+\.\d+)'

extracted = df_logs['Raw_Server_Log'].str.extract(pattern)
extracted.columns = ['Date', 'user_id', 'product_id', 'eur_val']

# Cast to correct types
extracted['Date'] = pd.to_datetime(extracted['Date'])
extracted['eur_val'] = pd.to_numeric(extracted['eur_val'])

# Drop any rows where regex found no match (safety net)
extracted_data = extracted.dropna().reset_index(drop=True)

print(f"✅ Phase 2 Complete: {len(extracted_data):,} clean transactions extracted")
print(f"   Date range: {extracted_data['Date'].min().date()} to {extracted_data['Date'].max().date()}")
extracted_data.head()
````

---
## PHASE 3: Financial Reconciliation — Temporal Join with Forward-Fill

**The Trap:** The exchange rate CSV has **no entries for weekends** (markets are closed). A standard `merge()` / `INNER JOIN` on Date would silently delete ~28% of transactions that fall on Saturdays and Sundays.

**The Solution:**  
1. Reindex the exchange rate table to a **complete daily date range** (no gaps).  
2. **Forward-fill (`ffill`)** — rolls Friday's rate into Saturday and Sunday.  
3. Then merge safely — every transaction date now has a valid rate.

````python
# ===================================================================
# PHASE 3: FINANCIAL RECONCILIATION & MERGING
# ===================================================================

# --- Step 1: Load exchange rates ---
df_rates = pd.read_csv('daily_exchange_rates.csv')
df_rates['Date'] = pd.to_datetime(df_rates['Date'])

missing_before = df_rates['EUR_to_USD'].isna().sum()
print(f"   Missing rates before ffill: {missing_before}")

# --- Step 2: Reindex to full daily range and forward-fill weekend gaps ---
# Create a complete daily date index spanning the whole year
full_date_range = pd.date_range(
    start=df_rates['Date'].min(),
    end=df_rates['Date'].max(),
    freq='D'
)
# Reindex introduces NaN for missing weekend dates, then ffill carries Friday forward
df_rates = (
    df_rates
    .set_index('Date')
    .reindex(full_date_range)
    .rename_axis('Date')
    .reset_index()
)
df_rates['EUR_to_USD'] = df_rates['EUR_to_USD'].ffill()
print(f"   Missing rates after ffill:  {df_rates['EUR_to_USD'].isna().sum()}")
print(f"   Exchange rate rows (full year): {len(df_rates)}")

# --- Step 3: Merge transactions with exchange rates (temporal join by Date) ---
df_merged = extracted_data.merge(df_rates, on='Date', how='left')
print(f"   After rate merge: {len(df_merged):,} rows")

# --- Step 4: Merge with active users (inner join keeps only active users) ---
df_merged = df_merged.merge(
    df_active_users[['user_id', 'name']],
    on='user_id',
    how='inner'   # Intentionally drops transactions from Suspended/Pending users
)
print(f"   After active-user filter: {len(df_merged):,} rows")

# --- Step 5: Calculate USD Revenue (vectorized — no loops) ---
df_merged['USD_Revenue'] = df_merged['eur_val'] * df_merged['EUR_to_USD']

# Final clean-up
df_final = df_merged[['Date', 'user_id', 'name', 'product_id',
                       'eur_val', 'EUR_to_USD', 'USD_Revenue']].copy()
df_final['Month'] = df_final['Date'].dt.to_period('M')

elapsed = time.time() - pipeline_start
print(f"\n✅ Phase 3 Complete: df_final has {len(df_final):,} rows")
print(f"   Total USD Revenue: ${df_final['USD_Revenue'].sum():,.2f}")
print(f"   ⏱  Full pipeline elapsed time: {elapsed:.2f}s (limit: 120s)")
df_final.head()
````

---
## PHASE 4: CFO Dashboard — Aggregation & Visualization

````python
# ===================================================================
# PHASE 4: AGGREGATION & VISUALIZATION (CFO DASHBOARD)
# ===================================================================

# --- Aggregation 1: Total USD Revenue by Month ---
monthly_revenue = (
    df_final
    .groupby('Month')['USD_Revenue']
    .sum()
    .reset_index()
)
monthly_revenue['Month_str'] = monthly_revenue['Month'].dt.strftime('%b %Y')
monthly_revenue['Month_num'] = monthly_revenue['Month'].apply(lambda x: x.month)

# --- Aggregation 2: Top 5 Most Valuable Customers ---
top5_customers = (
    df_final
    .groupby(['user_id', 'name'])['USD_Revenue']
    .sum()
    .nlargest(5)
    .reset_index()
)
top5_customers['Label'] = top5_customers['user_id']  # Short label for chart

# --- Dashboard Layout ---
sns.set_theme(style='whitegrid', palette='muted')
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Q1–Q4 Financial Audit Dashboard — 2023', fontsize=16, fontweight='bold', y=1.02)

# ---- Chart 1: Monthly Revenue Bar Chart ----
ax1 = axes[0]
bars = ax1.bar(
    monthly_revenue['Month_str'],
    monthly_revenue['USD_Revenue'],
    color=sns.color_palette('Blues_d', len(monthly_revenue)),
    edgecolor='white'
)
# Annotate bars with value labels
for bar in bars:
    height = bar.get_height()
    ax1.annotate(
        f'${height/1e6:.1f}M',
        xy=(bar.get_x() + bar.get_width() / 2, height),
        xytext=(0, 4), textcoords='offset points',
        ha='center', va='bottom', fontsize=8, fontweight='bold'
    )
ax1.set_title('Total USD Revenue by Month', fontsize=13, fontweight='bold')
ax1.set_xlabel('Month', fontsize=11)
ax1.set_ylabel('Revenue (USD)', fontsize=11)
ax1.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x/1e6:.1f}M'))
ax1.tick_params(axis='x', rotation=45)
ax1.set_ylim(0, monthly_revenue['USD_Revenue'].max() * 1.15)

# ---- Chart 2: Top 5 Customers Horizontal Bar Chart ----
ax2 = axes[1]
colors = sns.color_palette('Oranges_d', 5)
hbars = ax2.barh(
    top5_customers['Label'],
    top5_customers['USD_Revenue'],
    color=colors,
    edgecolor='white'
)
# Annotate with dollar values
for bar, val in zip(hbars, top5_customers['USD_Revenue']):
    ax2.text(
        val + top5_customers['USD_Revenue'].max() * 0.01,
        bar.get_y() + bar.get_height() / 2,
        f'${val:,.0f}',
        va='center', ha='left', fontsize=9, fontweight='bold'
    )
ax2.set_title('Top 5 Most Valuable Customers (Total USD)', fontsize=13, fontweight='bold')
ax2.set_xlabel('Total Revenue (USD)', fontsize=11)
ax2.set_ylabel('Customer ID', fontsize=11)
ax2.invert_yaxis()  # Highest revenue at the top
ax2.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax2.set_xlim(0, top5_customers['USD_Revenue'].max() * 1.25)

plt.tight_layout()
plt.savefig('cfo_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Phase 4 Complete: CFO Dashboard generated and saved as cfo_dashboard.png')
````

---
## Key Technical Decisions — Why This Code Passes

| Requirement | Technique Used | Why |
|---|---|---|
| **Latest user status** | `ROW_NUMBER() OVER (PARTITION BY user_id ORDER BY updated_at DESC)` | Window function assigns rank 1 to most recent row per user |
| **Extract from raw logs** | `df['col'].str.extract(pattern)` | Vectorized — applies regex to all 100k rows in one operation |
| **Drop ERROR rows** | `~df['col'].str.contains('ERROR')` | Vectorized boolean mask — no loop |
| **Weekend rate gaps** | `reindex(full_date_range)` + `ffill()` | Reindex creates NaN for missing dates; ffill rolls Friday forward |
| **Temporal join** | `merge(on='Date', how='left')` on the complete rate table | No rows dropped because every date now has a rate |
| **Active-user filter** | `merge(how='inner')` with `df_active_users` | Inner join naturally excludes Suspended/Pending users |
| **USD calculation** | `df['eur_val'] * df['EUR_to_USD']` | Vectorized element-wise multiplication — no loop |
